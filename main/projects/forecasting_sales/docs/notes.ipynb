¿Qué variables influyen en la venta general?

- Número de bicicletas disponibles: 5
- Número de bicicletas dañanas: 3
- Número de biciletas robadas: 1

- Número de membresias vendidas por tipo: ECOBICI+: 20, HSBC: 30, ANUAL: 200  
- Ingreso por tipo de membresia: ECOBICI+: $20,000, HSBC: $30,000, ANUAL: $200,000

- Usuarios activos: con viajes disponibles
- Usuarios nuevos: con nuevas suscripciones

- Día: si es entre semana o fín de semana, día festivo
- Mes del año: temporalidad festiva o clima
- Es quincena: 15 o 30 de cada mes

2. Para proyectar venta de Membresías (Suscripciones)
La tabla BikeSubscriptionFact (Sección 5.4) registra cuándo alguien compra un pase o membresía. Esto es vital porque es ingreso recurrente o anticipado.

Para la serie temporal:

- `purchaseDate_id` y `purchaseTime_id`: Fecha y hora exacta de la compra.

Para el valor de la venta:

- `purchasePrice`: El precio inicial de la suscripción.
- `totalPaid`: El precio final pagado después de descuentos o ajustes.

Para segmentar:

- `subscriptionType_id`: Crucial para saber qué producto se vende más (Pase diario vs. Membresía anual).
- `purchaseStation_id`: Dónde se compró (útil para saber qué quioscos venden más).
