```python
import pandas as pd
```

```python
csv_file_path = '../data/Bin_Manager_Anexo_ECOBICI.csv'
df_transactions = pd.read_csv(csv_file_path, encoding='latin1')
df_transactions.head()
```

```python
# quitar a BIN los ultimos dos digitos
df_transactions['BIN'] = df_transactions['BIN'].astype(str).str[:-2]
df_transactions.head()
```

```python
# contar duplicados
df_transactions.duplicated().sum()
```

```python
# eliminar duplicados
df_transactions = df_transactions.drop_duplicates()
df_transactions.head()
```

```python
# filtrar por BIN
bin_to_filter = '418914'  # Reemplaza con el BIN que deseas filtrar
filtered_df = df_transactions[df_transactions['BIN'] == bin_to_filter]
filtered_df.head()
```

```python
df_transactions.to_csv('../data/Bin_ECOBICI.csv', index=False)
```
